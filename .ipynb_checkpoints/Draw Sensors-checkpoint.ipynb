```python
import plotly.graph_objects as go
import plotly.io as pio
import pandas as pd
from plotly.subplots import make_subplots
```

```python
xls = pd.ExcelFile('df_radon_combined.xlsx')

xls.sheet_names 

df = {}
for num in xls.sheet_names[2:]:
    df[num] = pd.read_excel(xls, num) 
    print(df[num].shape) 

#remove these indexes
devices = ['1', '2', '5', '9', '17', '23', '27', '28', '36', '44', '48']
for key in list(df):
    if key in devices:
        del df[key]

for key in df:
    print(key)
    print(df[key].shape)
    df[key]['SyncDate'] = pd.to_datetime(df[key]['SyncDate'])
    print(f"{df[key]['SyncDate'].min()} - {df[key]['SyncDate'].max()}")
    df[key].sort_values(by='SyncDate', ascending=True, inplace=True)
    df[key]['SyncDate'] = df[key]['SyncDate'].dt.floor('H')
    df[key] = df[key].resample('H', on = 'SyncDate').mean()
    df[key] = df[key].interpolate(method='linear', limit_direction='both')
    start_time = pd.to_datetime('2022-05-11T18:29:00.000000000')
    end_time = pd.to_datetime('2023-06-06T12:00:00.000000000')
    df[key] = df[key][(df[key].index >= start_time) & (df[key].index <= end_time)]
    print(f"{df[key].index.min()} - {df[key].index.max()}")
```

```python
df_radon = pd.DataFrame(index=df['50'].index)
for key in df:
    print(key)
    df_radon[f'Radon_{key}'] = df[key]['Radon']
```

```python
df_radon
```

```python

# Create initial traces for all sensors
traces = [go.Scatter(x=df_radon.index, y=df_radon[sensor], mode='lines', name=sensor, visible='legendonly') for sensor in df_radon.columns]

# Create layout with dropdown menu
layout = go.Layout(
    title='Sensor Data',
    xaxis=dict(title='Time'),
    yaxis=dict(title='Sensor Value')
)

# Create subplot figure with initial data and layout
fig = make_subplots(rows=1, cols=1, shared_xaxes=True)
for trace in traces:
    fig.add_trace(trace, row=1, col=1)

# Create checkboxes for sensor selection
checkboxes = [
    dict(
        label=sensor,
        method='update',
        args=[{'visible': [sensor == trace.name or trace.visible == 'legendonly' for trace in fig.data]}]
    )
    for sensor in df_radon.columns
]

# Add checkboxes
fig.update_layout(
    updatemenus=[
        dict(
            buttons=[
                {'label': 'All', 'method': 'update', 'args': [{'visible': [True] * len(df_radon.columns)}]},
                {'label': 'None', 'method': 'update', 'args': [{'visible': [False] * len(df_radon.columns)}]}
            ],
            direction='down',
            showactive=True,
            x=0.01,
            xanchor='left',
            y=1.15,
            yanchor='top'
        )
    ],
    annotations=[
        dict(
            text="Select Sensors:",
            x=0,
            y=1.13,
            xref="paper",
            yref="paper",
            align="left",
            showarrow=False
        )
    ]
)

# Add checkboxes to annotations
for i, checkbox in enumerate(checkboxes):
    checkbox['y'] = 1.09 - i * 0.04
    checkbox['xref'] = 'paper'
    checkbox['yref'] = 'paper'
    fig.add_annotation(checkbox)

# Overlay all sensors
fig.update_traces(visible=True)

# Show the chart
pio.show(fig)
```
